```python
# Visualize the relationship between daily customer count and store sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Stores, x='Daily_Customer_Count', y='Store_Sales')
plt.title('Daily Customer Count vs. Store_Sales')
plt.xlabel('Daily Customer Count')
plt.ylabel('Store_Sales')
plt.show()
```

```python
# Visualize the relationship between item available and store sales

plt.figure(figsize=(10, 6))

sns.scatterplot(data=Stores, x='Items_Available', y='Store_Sales')

plt.title('Items Available vs. Store_Sales')

plt.xlabel('Items Available')

plt.ylabel('Store_Sales')
```

```python
## To get the store with the largest area

largest_area_store = Stores.loc[Stores['Store_Area'].idxmax()]
print(largest_area_store)
```

```python
## To get the store with the most daily customers

most_customers_store = Stores.loc[Stores['Daily_Customer_Count'].idxmax()]
print(most_customers_store)
```

```python
# Histogram of Daily Customer Count

plt.figure(figsize=(10, 6))
sns.histplot(data=Stores, x='Daily_Customer_Count', bins=20, kde=True)
plt.title('Histogram of Daily Customer Count')
plt.xlabel('Daily Customer Count')
plt.ylabel('Frequency')
plt.show()
```

## Observation

The histogram shows that the distribution of the Daily Customer Count. The x-axis represents the Daily Customer Counts, and the y-axis represents the frequency at which the sales (number of occurrences) of each interval (bin). The KDE curve provides a smooth estimate of the data's probability density function, allowing you to visualize the shape of the distribution more easily.

```python
# Sort the data by 'Store_Sales' in descending order and take the top 10 rows
top_10_sales = Stores.sort_values(by='Store_Sales', ascending=False).head(10)

# Create a figure for the plot with a specified size
plt.figure(figsize=(10, 6))

# Create a bar plot using Seaborn, where 'Store_Area' is on the x-axis and 'Store_Sales' on the y-axis
sns.barplot(data=top_10_sales, x='Store_Area', y='Store_Sales')

# Set the title of the plot
plt.title('Bar Plot of Top 10 Store Sales vs. Store Area')

# Label the x-axis
plt.xlabel('Store Area')

# Label the y-axis
plt.ylabel('Store Sales')

# Show the plot
plt.show()
```

```python
# Sales per square foot

Stores['Sales_per_sqft'] = Stores['Store_Sales'] / Stores['Store_Area']
Stores
```

```python
Stores_sorted = Stores.sort_values(by='Sales_per_sqft', ascending=False)
Stores_sorted
```

```python
top_10_stores = Stores_sorted.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_stores, x='Store ID ', y='Sales_per_sqft', color='blue')
plt.title('Top 10 Stores: Sales per Square Foot')
plt.xlabel('Store ID')
plt.ylabel('Sales per Square Foot')
plt.xticks(rotation=90)
plt.show()
```

```python
# Average daily sales
Stores['Avg_Daily_Sales'] = Stores['Store_Sales'] / Stores['Daily_Customer_Count']
Stores
```

```python
##  Items Available Distribution

top_5_items = Stores['Items_Available'].value_counts().head(5)

plt.figure(figsize=(8, 8))
plt.pie(top_5_items, labels=top_5_items.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Items Available Distribution')
plt.show()
```

```python
# Calculate the average store sales for each daily customer count

avg_sales = Stores.groupby('Daily_Customer_Count')['Store_Sales'].mean().reset_index()
Stores
```

```python
# Sort the dataframe by average store sales in descending order

avg_sales_sorted = avg_sales.sort_values('Store_Sales', ascending=False)
Stores
```

```python
# Create a bar chart to display the top 5 daily customer counts by average store sales

plt.figure(figsize=(8, 4))

sns.barplot(data=avg_sales_sorted.head(5), x='Daily_Customer_Count', y='Store_Sales')

plt.title('Top 5 Daily Customer Counts by Average Store Sales')

plt.xlabel('Daily Customer Count')

plt.ylabel('Average Store Sales')

plt.show()
```
